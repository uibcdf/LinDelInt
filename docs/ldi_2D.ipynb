```python
import lindelint as ldi
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([[-4.5, -6.8],
                   [ 7.5, -5.3],
                   [-4.1,  5.1],
                   [ 5.7,  4.7],
                   [-6.6,  0.2],
                   [ 2.9, -0.7]])

properties = np.array([[ 0.5, -0.5],
                       [-0.7, -0.9],
                       [ 0.0,  0.8],
                       [-0.6,  0.9],
                       [ 0.4,  0.3],
                       [-0.3, -0.5]])
```

```python
new_points = np.random.uniform(-10, 10, (400, 2))
```

```python
plt.rcParams["figure.figsize"] = (6,6)

ax=plt.gca()

for ii in range(new_points.shape[0]):
    plt.scatter(new_points[ii,0], new_points[ii,1], c='k', s=0.5)
    
for ii in range(points.shape[0]):
    plt.arrow(points[ii,0], points[ii,1], properties[ii,0], properties[ii,1],
              head_width=0.2, head_length=0.4, fc='r', ec='r', lw=2.0)

    
plt.axis('scaled')
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.show()
```

```python
interpolator = ldi.Interpolator(points, properties)
```

```python
new_properties = interpolator.do_your_thing(new_points)
```

```python
from IPython.core.display import HTML as Center

Center(""" <style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style> """)

plt.rcParams["figure.figsize"] = (6,6)

ax=plt.gca()

for ii in range(new_points.shape[0]):
    plt.arrow(new_points[ii,0], new_points[ii,1], new_properties[ii,0], new_properties[ii,1],
        head_width=0.2, head_length=0.4, fc='k', ec='k')
    
for ii in range(points.shape[0]):
    plt.arrow(points[ii,0], points[ii,1], properties[ii,0], properties[ii,1],
              head_width=0.2, head_length=0.4, fc='r', ec='r', lw=2.0)
    
plt.axis('scaled')
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.show()
```
